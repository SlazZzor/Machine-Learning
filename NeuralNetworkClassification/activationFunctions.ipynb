```python
import torch
import matplotlib.pyplot as plt
A = torch.arange(-10, 10, 1, dtype=torch.float32)
A
```

```python
plt.plot(A);
```

```python
def relu(x):
    return torch.maximum(torch.tensor(0), x)
relu(A)
```

```python
plt.plot(relu(A))
```

```python
def sigmoid(x):
    return 1 / (1 + torch.exp(-x))

sigmoid(A)
```

```python
plt.plot(sigmoid(A))
```
